```python
nfl_main_df['date'] = pd.to_datetime(nfl_main_df['date'], format='%Y%m%d')
```

```python
nfl_main_df['season'] = nfl_main_df['date'].dt.isocalendar().year
nfl_main_df['week'] = nfl_main_df['date'].dt.isocalendar().week
nfl_main_df['day'] = nfl_main_df['date'].dt.isocalendar().day
```

```python
nfl_main_df = nfl_main_df.drop(nfl_main_df[nfl_main_df.week < 36].index)
```

```python
nfl_main_df["week"] = nfl_main_df["week"]-34
```

```python
nfl_main_df["week"] = nfl_main_df["week"]-1
```

```python
nfl_main_df["week"] = np.where(nfl_main_df["day"] == 1, nfl_main_df["week"]-1, nfl_main_df["week"])
nfl_main_df["week"] = np.where(nfl_main_df["day"] == 2, nfl_main_df["week"]-1, nfl_main_df["week"])
nfl_main_df["day"] = np.where(nfl_main_df["day"] == 1, "Mon", nfl_main_df["day"])
nfl_main_df["day"] = np.where(nfl_main_df["day"] == 2, "Tue", nfl_main_df["day"])
nfl_main_df["day"] = np.where(nfl_main_df["day"] == 3, "Wed", nfl_main_df["day"])
nfl_main_df["day"] = np.where(nfl_main_df["day"] == 4, "Thu", nfl_main_df["day"])
nfl_main_df["day"] = np.where(nfl_main_df["day"] == 5, "Fri", nfl_main_df["day"])
nfl_main_df["day"] = np.where(nfl_main_df["day"] == 6, "Sat", nfl_main_df["day"])
nfl_main_df["day"] = np.where(nfl_main_df["day"] == 7, "Sun", nfl_main_df["day"])
```

```python
nfl_main_df["away_spread_open"] = np.where(nfl_main_df["away_spread_open"] == 'PK', "0", nfl_main_df["away_spread_open"])
nfl_main_df["away_spread_close"] = np.where(nfl_main_df["away_spread_close"] == 'PK', "0", nfl_main_df["away_spread_close"])
nfl_main_df["home_spread_open"] = np.where(nfl_main_df["home_spread_open"] == 'PK', "0", nfl_main_df["home_spread_open"])
nfl_main_df["home_spread_close"] = np.where(nfl_main_df["home_spread_close"] == 'PK', "0", nfl_main_df["home_spread_close"])
nfl_main_df["away_moneyline_open"] = np.where(nfl_main_df["away_moneyline_open"] == '-', "0", nfl_main_df["away_spread_open"])
```

```python
column_to_move = nfl_main_df.pop("season")
nfl_main_df.insert(2, "season", column_to_move)
column_to_move = nfl_main_df.pop("week")
nfl_main_df.insert(3, "week", column_to_move)
column_to_move = nfl_main_df.pop("day")
nfl_main_df.insert(4, "day", column_to_move)
```

```python
byes_df = (sqldf('''SELECT a.Season, a.Week, a.AwayTeam as Team
                      FROM df as a 
                      WHERE a.Week > 3
                      AND NOT EXISTS
                          (SELECT *
                            FROM df as b
                            WHERE a.Season = b.Season
                            and a.Week = b.Week - 1
                            and (a.AwayTeam = b.AwayTeam or a.AwayTeam = b.HomeTeam) 
                            )
                UNION
                SELECT a.Season, a.Week, a.HomeTeam as Team
                      FROM df as a 
                      WHERE a.Week > 3
                      AND NOT EXISTS
                          (SELECT *
                            FROM df as b
                            WHERE a.Season = b.Season
                            and a.Week = b.Week - 1
                            and (a.HomeTeam = b.AwayTeam or a.HomeTeam = b.HomeTeam) 
                            )
                     '''))
```

```python
results_df = (sqldf('''SELECT a.GameID
                    , a.Season
                    , a.Week
                    , a.Day
                    , a.AwayTeam
                    , 'W' as AwayTeamResult
                    , a.AwayScore
                    , a.AwaySpreadClose as SpreadClose
                    , a.HomeScore
                    , 'L' as HomeTeamResult
                    , a.HomeTeam
                    , a.AwayScore + a.AwaySpreadClose - a.HomeScore as WinMargin
                FROM df as a
                WHERE ( a.AwayScore + a.AwaySpreadClose > a.HomeScore + 0)
                UNION
                SELECT a.GameID
                    , a.Season
                    , a.Week
                    , a.Day
                    , a.AwayTeam
                    , 'L' as AwayTeamResult
                    , a.AwayScore
                    , a.AwaySpreadClose as SpreadClose
                    , a.HomeScore
                    , 'W' as HomeTeamResult
                    , a.HomeTeam
                    , a.HomeScore - a.AwaySpreadClose - a.AwayScore
                FROM df as a
                WHERE ( a.HomeScore - a.AwaySpreadClose > a.AwayScore + 0)
                UNION
                SELECT a.GameID
                    , a.Season
                    , a.Week
                    , a.Day
                    , a.AwayTeam
                    , 'P' as AwayTeamResult
                    , a.AwayScore
                    , a.AwaySpreadClose as SpreadClose
                    , a.HomeScore
                    , 'P' as HomeTeamResult
                    , a.HomeTeam
                    , 0.0
                FROM df as a
                WHERE ( a.HomeScore - a.AwaySpreadClose = a.AwayScore + 0)
                ORDER BY a.GameID
            '''))
results_df
```
